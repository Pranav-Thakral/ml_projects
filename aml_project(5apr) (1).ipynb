```python
#outlier plotting
```

```python
df.plot.scatter(x='BMI', y='HeartDisease', color='blue', marker='o', s=100, label='Data Points')
plt.xlabel('BMI')
plt.ylabel('HeartDisease')
plt.title('Scatter Plot (HeartDIsease and BMI')
plt.show()
```

```python
df.plot.scatter(x='BMI', y='PhysicalHealth', color='blue', marker='o', s=70)
plt.xlabel('BMI')
plt.ylabel('PhysicalHealth')
plt.title('Scatter Plot (BMI and PhysicalHealth)')
plt.show()
```

```python
df.plot.scatter(x='BMI', y='SleepTime', color='blue', marker='o', s=80, label='Data Points')
plt.xlabel('BMI')
plt.ylabel('SleepTime')
plt.title('Scatter Plot (BMI and SleepTime)')
plt.show()
```

```python
df.plot.scatter(x='BMI', y='MentalHealth', color='blue', marker='o', s=100)
plt.xlabel('BMI')
plt.ylabel('MentalHalth')
plt.title('Scatter Plot (BMI and MentalHealth)')
plt.show()
```

```python
df.plot.scatter(x='SleepTime', y='HeartDisease', color='blue', marker='o', s=100)
plt.xlabel('SleepTime')
plt.ylabel('HeartDisease')
plt.title('Scatter Plot (Sleeptime and HeartDisease)')
plt.show()
```

```python
#Histogram of categorical columns
```

```python
plt.figure(figsize=(10, 4))
sns.histplot(data=df, x='BMI', bins=20, kde=True)
plt.title('Histogram of BMI Column')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.histplot(data=df, x='PhysicalHealth', bins=20, kde=True)
plt.title('Histogram of PhysicalHealth Column')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.histplot(data=df, x='MentalHealth', bins=20, kde=True)
plt.title('Histogram of MentalHealth Column')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.histplot(data=df, x='SleepTime', bins=20, kde=True)
plt.title('Histogram of SleepTime Column')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
#countplot of categorical columns
```

```python
plt.figure(figsize=(10, 4))
sns.countplot(data=df, x='BMI')
plt.title('Count Plot of BMI Column')
plt.xlabel('values')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
